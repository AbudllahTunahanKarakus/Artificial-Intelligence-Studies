```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

dosya_yolu="/content/iris.csv"
iris=pd.read_csv(dosya_yolu)
# x özellikler y hedef değişken
X=iris.drop(["Id","Species"],axis=1) #ID VE Species kolonlarını cıkarıyoruz sadece sayısal özellikler kalıyor
y=iris["Species"] # hedef değişken turlerımız (setosa,veersicolor,virginica)
#veri setini eğtim ve test için %70 %30 oranın da ayırıryoruz
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=0)
model=RandomForestClassifier() # random forest sınıflandırıcıyı kullanıyoruz
model.fit(X_train,y_train)#modeli eğtim verisi ile eğitiyoruz

y_pred=model.predict(X_test) # test verisi ile tahmın yapıyoruz

#model doğruluk oranı hesaplıyoz
doğruluk=accuracy_score(y_test,y_pred)
print("Doğruluk Oranı:",doğruluk)
#modelin sınıflara gore detaylı performans raporunu yazdırıyoruz
rapor=classification_report(y_test,y_pred)
print("Sınıflandırma Raporu: \n",rapor)
```
