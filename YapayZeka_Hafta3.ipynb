```python
def OrtalamaIleDoldur(data):
  toplam=data.sum().sum()
  elemanSayısı=data.size - data.isnull().sum().sum()
  ortalama=toplam/elemanSayısı
  return ortalama # toplam Nan olmayan hucre sayısını buluruz tum değerler toplamına boleriz ortalamaya atarız  bu fonksıyon da
```

```python
data.fillna(OrtalamaIleDoldur(data)) # fonksyon ile bulann a ortalamayı Nan değerleri ile doldururuz
```

```python
data.mean(skipna=True) #sayısal kolonların ortalamsını hesaplar  skipna ile nan değerler atlanır  her sutunun ort değeri karsına yazılır nansız hesaplanan
```

```python
data.fillna(value=data.mean(skipna=True)) # her sayısal kolonun nan değerleri atlanarak ortalama hesaplanır value ile nan olanlara o sutnun ortalamsının değeri atanr
```
