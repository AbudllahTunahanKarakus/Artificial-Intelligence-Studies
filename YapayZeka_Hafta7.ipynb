```python
dosya_yolu="/content/data.csv"
data=pd.read_csv(dosya_yolu)
data.head()
data.diagnosis=[1 if i=="M" else 0 for i in data.diagnosis] # M VE BLERI KENDIMIZ 1 0 YAPABILDIK

data.drop(["id","Unnamed: 32"],axis=1,inplace=True)

y=data.diagnosis.values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)

knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
sonuc=knn.score(X_test,y_test)
print("KNN Doğruluk Oranı:",sonuc)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
acc=accuracy_score(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print(cm)
print(acc)
print(cr)
```

```python
from sklearn.model_selection import cross_val_score,KFold

dosya_yolu="/content/data.csv"
data=pd.read_csv(dosya_yolu)
data.head()
data.diagnosis=[1 if i=="M" else 0 for i in data.diagnosis] # M VE BLERI KENDIMIZ 1 0 YAPABILDIK

data.drop(["id","Unnamed: 32"],axis=1,inplace=True)

y=data.diagnosis.values

# 2. MODEL TANIMI
# K=11 komşulu bir KNN modeli kuruyoruz.
knn=KNeighborsClassifier(n_neighbors=11)
# 3. K-FOLD AYARLARI (SINAV SİSTEMİ)
# n_splits=10: Veriyi 10 eşit parçaya böl.
# shuffle=True: Bölmeden önce veriyi iyice karıştır (Sıralı gelirse hata olmasın diye).
# random_state=42: Her çalıştırdığında aynı karışıklığı versin.
kf=KFold(n_splits=10,shuffle=True,random_state=42)
# 4. ÇAPRAZ DOĞRULAMA (CROSS VALIDATION) ÇALIŞTIRMA
# Bu fonksiyon otomatik olarak bir döngü kurar:
# 1. Tur: 1. parçayı test, kalan 9 parçayı eğitim yapar -> Puanı not eder.
# 2. Tur: 2. parçayı test, kalanları eğitim yapar -> Puanı not eder.
# ... Bu işlem 10 kez tekrarlanır.
cv_sonuçları=cross_val_score(knn,X,y,cv=kf,scoring="accuracy")
print("her fold için accuarcy:",cv_sonuçları)
print("Ortalama accuracy:",cv_sonuçları.mean())
print("Standart sapma:",cv_sonuçları.std())
```
