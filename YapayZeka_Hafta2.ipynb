```python
import numpy as np
```

```python
data=pd.DataFrame(np.random.randn(6,6), # 6x6 normal dağılımlı 0 1 rastgele sayı retır 2 boyutlu numpy array
                  index=["Mete","Ferhat","Buket","Batu","Rabiye","Buğrahan"],
                  columns=["CRP","ALT","AST","P","Na","Glukoz"])
data
```

```python
data.loc["Buket" : "Rabiye","ALT" :"Na"]=np.nan # buketle rabıa arası Alt ile Na arası  seçilen satır ve stünü tum değerleri NaN yapar
data
```

```python
data.isnull().sum() # hangi kolonlar da eksik değer sayısını verir
```

```python
data.isnull().sum().sum() #toplam kaç veri eksiği oldugunu bulur
```

```python
data.info()
```

```python
data.fillna(value=0) #fillna eksik verileri doldurur eksik değerler yerine 0 koyar bu komut orj datayı etkılemez
```

```python
data
```

```python
data.fillna(value=1,inplace=True ) # Nan yerıne 1 yazar kalıcı olur
```

```python
data
```

```python
data.loc["Buket":"Rabiye","ALT":"Na"]=np.nan #
data
```

```python
data
```

```python
data.fillna(method="ffill") #nan deperini doldurur forward fill yontemi her eksık verinin yerini üstğndeki değerle doldurur kalıcı değil
```

```python
data.fillna(method="bfill") # bacward fill metodu her nan değerini altında kı değer ile dolddurur
```

```python
data
```

```python
data.fillna(value=data.mean()) # data mean her kolonun ort hesaplar Nan değeri yerıne her kolon kendı ortalamsı ile doldurur orj dadtayı etkilemez
```

```python
data.fillna(value=data.median()) #her kolunun medyan değerini bulur Nan olanları doldurur
```

```python
data.fillna(value=data.mean(),inplace=True)
data # aynısı kalıcıc olur
```
